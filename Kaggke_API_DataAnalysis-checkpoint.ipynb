```python
import pandas as pd
```

```python
# making not avialble and unknown data as null while reading data
df=pd.read_csv("orders.csv",na_values=['Not Available','unknown'])
```

```python
df.head(5)
```

```python
df['Ship Mode'].unique()
```

```python
#renaming the column name
# one way - df.rename(columns={'Order Id':'o'order_id','City:'city'})
df.columns
```

```python
#making the column names into lower  case and replacing ' ' as '_'
df.columns=df.columns.str.lower()
```

```python
df.columns=df.columns.str.replace(' ','_')
```

```python
df.head(5)
```

```python
df.columns
```

```python
#derive new columns discount , sale price and profit 

df['discount']=df['list_price']*df['discount_percent']*.01

df['sale_price']=df['list_price']-df['discount']

df['profit']=df['sale_price']-df['cost_price']
```

```python
df.head(5)
```

```python
#convert order date from object data type to datetime
df.dtypes
```

```python
df['order_date']=pd.to_datetime(df['order_date'],format="%Y-%m-%d")
```

```python
df.dtypes
```

```python
#droppping the columns which are not required
df.drop(columns=['cost_price','list_price','discount_percent'],inplace=True)
```
